```python
#pre-class conversations: python resources, pytest, URLLIB2|RE webscripting regular expressions (I missed it.)
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
#seaborn is a visualization package built on top of matplotlib to make the learning curve shorter
import seaborn as sns
#sets parameters of seaborn
sns.set()
#tells jupyter notebook to show plots in the notebook
%matplotlib inline
```

```python
wage_df = pd.read_csv("/Users/jason/Documents/pyplay/Wage.csv")
wage_df.head(n=2)
```

```python
# Making simple scatter plots with matplotlib

wage_df.plot(x = "age", y = "wage", kind = "scatter")
```

```python
wage_df.plot(x = "age", y = "wage", kind = "scatter",figsize = (10,8),title = "Relationship between age and wage")
```

```python
#google "seaborn" produces a gallery of seaborn type graphs
fig = sns.pairplot(wage_df,x_vars = "age", y_vars = "wage",kind = "reg")
```

```python
# Fiure aesthetics can be controlled through sns.set_style(). Five different background grids are available.
#sns.set_style("whitegrid")
sns.set_style("ticks")
#sns.despine(offset = 10, trim = True)
fig = sns.pairplot(wage_df,x_vars = "age", y_vars = "wage",kind = "reg")
sns.despine(trim = True, top = False, right = False) # only makes sense to use this after you make the plot !

```

```python
sns.set_style("darkgrid",{"axes.facecolor" : "0.4"})
#sns.despine(offset = 10, trim = True)
fig = sns.pairplot(wage_df,x_vars = "age", y_vars = "wage",kind = "reg")
#x and y vars is a tuple vs. a mutable object
```

```python
# Figure scale can be conotrolled using sns.set_context. Four contexts in order to size: paper, notebook, talk, poster

#"set_style" refers to style of the figure
sns.set_style("darkgrid",{"axes.facecolor" : "0.9"})
#"ste_context" refers to size of the figure
sns.set_context("poster",font_scale = 1.0)
#sns.despine(offset = 10, trim = True)
sns.pairplot(wage_df,x_vars = "age", y_vars = "wage",
             size = 5,aspect = 1.5,plot_kws = {'s':160})
#aspect is not a tuple and is mutable 
#'s' is the dot size
```
